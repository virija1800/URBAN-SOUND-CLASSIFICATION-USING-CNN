```python
import sys
import os
import IPython as IP
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display
import pickle
import wave
import gc
import tensorflow as tf
import matplotlib.pyplot as plt
from tqdm import tqdm
from pathlib import Path
from tensorflow.keras import layers
from tensorflow.python.client import device_lib
```

```python
audio_path = '/content/drive/MyDrive/archive'
metadata_path = '/content/drive/MyDrive/archive/UrbanSound8K.csv'
```

```python
metadata = pd.read_csv(metadata_path)
metadata.head(10)
```

```python
print("Number of audio files in entire dataset = ",len(metadata))
```

```python
#Defining some variables that will be needed.
features = []
labels = []
frames_max = 0
counter = 0
total_samples = len(metadata)
n_mfcc = 40
mfcc_max_padding=0
```

```python
#Iterating through all audio files and extracting MFCC 
#Time taken by this cell to execute = 1 hour 24 minutes
for index, row in metadata.iterrows():
    file_path = os.path.join(os.path.abspath(audio_path), 'fold' + str(row["fold"]), str(row["slice_file_name"]))
    class_label = row["class"]
    try:
        y, sr = librosa.load(file_path)
        normalized_y = librosa.util.normalize(y)
        mfcc = librosa.feature.mfcc(y=normalized_y, sr=sr, n_mfcc=n_mfcc)
        normalized_mfcc = librosa.util.normalize(mfcc)
        shape = normalized_mfcc.shape[1]
        mfccs = normalized_mfcc
        num_frames = mfccs.shape[1]
        features.append(mfccs)
        labels.append(class_label)
        if (num_frames > frames_max):
            frames_max = num_frames
        if (counter == 1):
            print("Status: {}/{}".format(index+1, total_samples))
            counter = 0
        counter += 1
    except Exception:
        pass
print("Finished: {}/{}".format(index, total_samples))
```

```python
# Given an numpy array of features, zero-pads each ocurrence to max_padding
def add_padding(features, mfcc_max_padding=174):
    padded = []
    for i in range(len(features)):
        px = features[i]
        size = len(px[0])
        # Add padding if required
        if (size < mfcc_max_padding):
            xDiff = mfcc_max_padding - size
            xLeft = xDiff//2
            xRight = xDiff-xLeft
            px = np.pad(px, pad_width=((0,0), (xLeft, xRight)), mode='constant')
        
        padded.append(px)

    return padded
```

```python
padded_features = add_padding(features, frames_max)
```

```python
print("Raw features length: {}".format(len(features)))
print("Padded features length: {}".format(len(padded_features)))
print("Feature labels length: {}".format(len(labels)))
```

```python
X = np.array(padded_features)  #Padded Feature are converted to numpy array & stored in X
y = np.array(labels)  #Labels are coverted to numpy array & stored in y
```

```python
np.save("/content/drive/MyDrive/archive/X-mfcc", X)
np.save("/content/drive/MyDrive/archive/y-mfcc", y)
```

```python
metadata = pd.read_csv("/content/drive/MyDrive/archive/UrbanSound8K.csv")
audio_data_directory = "/content/drive/MyDrive/archive/"
spectrograms_directory = "/content/drive/MyDrive/archive/Spectrograms/"   #Defining path to directory where spectrograms will be saved
```

```python
for i in range(0,10):
    Path(spectrograms_directory + "class_" + str(i)).mkdir(parents=True, exist_ok=True)
```

```python
def generate_spectrograms(row):

    audio_class = row["classID"]
    spect_directory = "class_" + str(audio_class)
    audio_file_name_without_extension = row["slice_file_name"][:-4]
    
    try:
      y, sr = librosa.load(audio_data_directory + "fold" + str(row["fold"]) + "/" + row["slice_file_name"])
      spectrogram = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)
      librosa.display.specshow(spectrogram, y_axis='linear')
      plt.savefig(spectrograms_directory + spect_directory + "/" + audio_file_name_without_extension + ".png")
      plt.clf() 
      plt.close('all')
      gc.collect()
    except:
      pass
```

```python
metadata.progress_apply(generate_spectrograms, axis=1)
```
